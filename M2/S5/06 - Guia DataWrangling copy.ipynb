El último número de la tarjeta es un dígito verificador y se calcula utilizando el algoritmo de Luhn. Se utiliza como medida de seguridad. La siguiente función permite verificar si un número de tarjeta es válido. 

Acá puedes obtener más información de cómo funciona el algoritmo de Luhn.

https://www.quobit.mx/asi-funciona-el-algoritmo-de-luhn-para-generar-numeros-de-tarjetas-de-credito.html

Cree una columna "valid" con valores True o False, en donde se determine si la tarjeta es válida o inválida


```python
def luhn_check(card_number):
    digits = [int(d) for d in str(card_number)]
    return sum(digits[::-2] + [sum(divmod(d * 2, 10)) for d in digits[-2::-2]]) % 10 == 0
```
