# --- 1: Introducción ---
La distribución muestral y el teorema del límite central son conceptos fundamentales en estadística y ciencia de datos. Estos tópicos permiten entender cómo se comportan las muestras extraídas de una población y cómo inferir propiedades de la población a partir de muestras.

# --- 2: La Distribución de Muestreo - Definición ---
La **distribución de muestreo** es la distribución de probabilidad de un estadístico (como la media, la proporción o la varianza) calculado a partir de múltiples muestras de una población.

## Ejercicio Explicativo 2: Identificando Estadísticos
Supongamos que estamos interesados en la altura promedio de los estudiantes universitarios. Recopilamos 1000 muestras, cada una de 50 estudiantes, y calculamos la altura promedio de cada una de esas 1000 muestras.
**Pregunta:** ¿Cuál es el estadístico cuya distribución de muestreo estamos creando?
**Respuesta:** La media muestral (X̄).

# --- 3: Media de la Distribución y Error Estándar ---
La media de la distribución muestral es igual a la media de la población: $E(verline{X}) = u$.
La desviación estándar de la distribución muestral (error estándar) se calcula como: $igma_{verline{X}} = igma/qrt{n}$, donde $igma$ es la desviación estándar poblacional y $n$ el tamaño de la muestra.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Ejercicio Explicativo 3: Cálculo del Error Estándar
mu_pop = 3000                # media poblacional
sigma_pop = 500              # desviación estándar poblacional
n = 100                      # tamaño de la muestra
e_x_bar = mu_pop             # media de la distribución muestral
sigma_x_bar = sigma_pop / np.sqrt(n)  # error estándar de la media muestral
print(f'Media de la distribución muestral: {e_x_bar}')
print(f'Error estándar: {sigma_x_bar}')
```

# --- 4: Ejemplo de Distribución Muestral ---
Supongamos que la población tiene una media $u=50$ y una desviación estándar $igma=10$. Si tomamos muestras de tamaño $n=25$, la distribución de muestreo de la media tendrá:
- Media: $E(verline{X}) = 50$
- Error estándar: $igma_{verline{X}} = 10/qrt{25} = 2$

Cuanto mayor es el tamaño de la muestra $n$, menor es el error estándar, lo que significa que las medias muestrales estarán más concentradas alrededor de la media poblacional.

```python
mu_pop_ex = 50                               # media poblacional de ejemplo
sigma_pop_ex = 10                            # desviación estándar poblacional de ejemplo
n_ex = 25                                    # tamaño de muestra inicial
mean_dist_ex = mu_pop_ex                     # media de la distribución muestral
se_dist_ex = sigma_pop_ex / np.sqrt(n_ex)    # error estándar para n=25
print(f'Media: {mean_dist_ex}')
print(f'Error estándar: {se_dist_ex}')
# ¿Qué pasa si n=100?
n_new = 100                                  # nuevo tamaño de muestra
se_new = sigma_pop_ex / np.sqrt(n_new)       # error estándar para n=100
print(f'Error estándar para n=100: {se_new}')
```

# --- 5: Ley de los Grandes Números - Definición y Fórmula ---
La Ley de los Grandes Números (LGN) establece que, a medida que el tamaño de la muestra $n$ aumenta, la media muestral $verline{X}$ se aproxima a la media poblacional $u$.

Si $X_1, X_2, ..., X_n$ son variables aleatorias independientes e idénticamente distribuidas con media $u$, entonces $verline{X} = (1/n) um X_i 	o u$ cuando $n 	o nfty$.

## Ejercicio Explicativo 5: Concepto de Convergencia
Imagina que lanzas un dado justo (caras 1 a 6). La media real es 3.5.
**Pregunta:** Si lanzas el dado solo 5 veces, ¿esperarías que el promedio sea exactamente 3.5? ¿Qué pasaría si lo lanzas 1000 veces?
**Respuesta:** Con solo 5 lanzamientos, el promedio puede diferir mucho de 3.5. Con 1000 lanzamientos, el promedio se acercará mucho a 3.5.

# --- 6: Ley de los Grandes Números - Tipos ---
- Ley Débil: Garantiza que la media muestral converge en probabilidad a la media poblacional.
- Ley Fuerte: Afirma que la media muestral converge casi seguramente a la media poblacional.

La Ley Fuerte implica una garantía más robusta: la media muestral eventualmente se mantendrá cerca de la media poblacional con probabilidad 1.

# --- 7: Ejemplos de la Ley de los Grandes Números ---
- Si lanzamos una moneda 10 veces, la proporción de caras puede ser diferente de 0.5.
- Si lanzamos la moneda 10,000 veces, la proporción de caras estará muy cerca de 0.5.
- Si medimos la altura de 5 personas, la media puede diferir de la media poblacional.
- Si medimos la altura de 100,000 personas, el promedio se acercará a la media real de la población.

```python
# Simulación de la LGN (Lanzamiento de Moneda)
n_lanzamientos = 10000
resultados = np.random.randint(0, 2, size=n_lanzamientos) # 0 cruz, 1 cara
proporciones_acumuladas = np.cumsum(resultados) / np.arange(1, n_lanzamientos + 1)
plt.figure(figsize=(10, 6))
plt.plot(proporciones_acumuladas, label='Proporción de caras acumulada')
plt.axhline(0.5, color='r', linestyle='--', label='Valor real (0.5)')
plt.xlabel('Número de Lanzamientos')
plt.ylabel('Proporción de Caras')
plt.title('Simulación de la Ley de los Grandes Números (Lanzamiento de Moneda)')
plt.legend()
plt.grid(True)
plt.show()
```

# --- 8: Teorema del Límite Central - Definición y Aplicación ---
El Teorema del Límite Central (TLC) afirma que, bajo ciertas condiciones, la distribución de la media muestral se aproxima a una distribución normal, sin importar la forma original de la población.
Esto ocurre cuando las variables son independientes e idénticamente distribuidas y el tamaño de muestra es suficientemente grande.

# --- 9: Ejemplos del Teorema del Límite Central ---
- Si el tiempo de espera en un restaurante tiene media 15 y varianza 9, al tomar muestras de n=30, la distribución de medias muestrales se aproxima a N(15, 0.3).
- Si la distribución original de alturas es asimétrica, para muestras de n=50, la distribución de medias muestrales será aproximadamente normal.

```python
# Simulación del TLC (Distribución Exponencial)
num_samples = 1000                  # número de muestras a simular
sample_size = 30                    # tamaño de cada muestra
lambda_exp = 0.5                    # parámetro lambda de la exponencial
sample_means = []                   # lista para guardar las medias muestrales
for _ in range(num_samples):
    sample = np.random.exponential(scale=1/lambda_exp, size=sample_size)  # generar muestra exponencial
    sample_means.append(np.mean(sample))                                  # calcular y guardar la media
import seaborn as sns
sns.histplot(sample_means, kde=True, stat='density')
plt.xlabel('Media Muestral')
plt.ylabel('Densidad')
plt.title(f'Distribución de Medias Muestrales (n={sample_size}) de una Distribución Exponencial')
plt.grid(True)
plt.show()
```

# --- 10: Relación entre Distribuciones ---
- **Distribución de la Población:** Todos los valores posibles de una variable en la población.
- **Distribución de la Muestra:** Valores de la variable en una muestra específica.
- **Distribución Muestral:** Distribución de un estadístico (media, proporción) calculado a partir de muchas muestras.

# --- 11: Cálculo de Probabilidades con la Distribución Muestral ---
Para la media muestral: $SE = igma/qrt{n}$.
Para la proporción muestral: $SE = qrt{p(1-p)/n}$.
Convertimos el valor de interés a una puntuación Z: $Z = (verline{X} - u)/(igma/qrt{n})$.
Luego usamos tablas Z o software estadístico para encontrar la probabilidad.

```python
# Ejemplo: Probabilidad con Z-score
# Parámetros poblacionales y de la muestra
mu = 2.5            # media poblacional
sigma = 0.1         # desviación estándar poblacional
n = 49              # tamaño de la muestra
x_bar_value = 2.52  # valor observado de la media muestral

# Cálculo del error estándar
se = sigma / np.sqrt(n)

# Cálculo del Z-score para la media muestral observada
z_score = (x_bar_value - mu) / se

# Cálculo de la probabilidad de obtener una media muestral mayor que 2.52
from scipy.stats import norm
probability = 1 - norm.cdf(z_score)
print(f'Z-score: {z_score:.2f}')
print(f'Probabilidad de que la media muestral sea mayor que 2.52 cm: {probability:.4f}')
```

# --- 12: Distribución Muestral de Proporciones ---
La distribución muestral de proporciones describe cómo varía la proporción muestral $at{p}$ en múltiples muestras.
Si $p$ es la proporción poblacional y se toman muestras de tamaño $n$, entonces $at{p}$ sigue aproximadamente una distribución normal cuando $np eq 5$ y $n(1-p) eq 5$.
La media es $E(at{p}) = p$ y el error estándar es $igma_{at{p}} = qrt{p(1-p)/n}$.

```python
# Ejemplo: Proporciones y condiciones
p_pop = 0.60                                  # proporción poblacional
n_prop = 80                                   # tamaño de la muestra
mean_p_hat = p_pop                            # media de la proporción muestral
se_p_hat = np.sqrt(p_pop * (1 - p_pop) / n_prop)  # error estándar de la proporción muestral
print(f'Media de p̂: {mean_p_hat}')           # muestra la media de p̂
print(f'Error estándar de p̂: {se_p_hat:.4f}')# muestra el error estándar de p̂
print(f'np = {n_prop * p_pop}')               # muestra el valor de np
print(f'n(1-p) = {n_prop * (1 - p_pop)}')     # muestra el valor de n(1-p)
if n_prop * p_pop >= 5 and n_prop * (1 - p_pop) >= 5:  # condición para aproximación normal
    print('Se puede usar la aproximación normal.')
else:
    print('No se puede usar la aproximación normal.')
```
