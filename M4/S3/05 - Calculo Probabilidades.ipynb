# Cálculo de Probabilidades en Python

```python
import numpy as np
import matplotlib.pyplot as plt
# importamos librería scipy
from scipy.stats import norm
```

## Caso 1
Supongamos que la puntuación en un examen de matemáticas sigue una distribución normal con una media de 75 y una desviación estándar de 10. Queremos calcular la probabilidad de obtener una puntuación mayor o igual a 85 en el examen.

```python
# Definir los parámetros de la distribución normal
mu = 75
sigma = 10

# Calcular la probabilidad de obtener una nota mayor o igual a 85 en el examen
probabilidad = 1 - norm.cdf(85, mu, sigma)

# Mostrar la probabilidad
print("La probabilidad de obtener una nota mayor o igual a 85 en el examen es:", round(probabilidad, 4))
```

```python
# Crear un arreglo de valores para la variable aleatoria
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# Calcular la función de densidad de probabilidad
pdf = norm.pdf(x, mu, sigma)

# Graficar la distribución normal
plt.plot(x, pdf, label='Distribución normal')

# Sombras para el intervalo de interés
x_fill = np.linspace(85, 100, 100)
y_fill = norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill, color='grey', alpha=0.5)

# Configurar la gráfica
plt.title('Distribución normal')
plt.xlabel('Nota')
plt.ylabel('Densidad de probabilidad')
plt.legend()

# Mostrar la gráfica
plt.show()
```

## Caso 2
Supongamos que la altura de una población de hombres adultos sigue una distribución normal con una media de 175 cm y una desviación estándar de 7 cm. Queremos calcular la probabilidad de que un hombre adulto seleccionado al azar tenga una altura entre 170 cm y 180 cm.

```python
# Definir los parámetros de la distribución normal
mu = 175
sigma = 7

# Calcular la probabilidad de que la altura esté entre 170 cm y 180 cm
probabilidad = norm.cdf(180, mu, sigma) - norm.cdf(170, mu, sigma)

# Mostrar la probabilidad
print("La probabilidad de que un hombre adulto tenga una altura entre 170 cm y 180 cm es:", round(probabilidad, 4))
```

```python
# Crear un arreglo de valores para la variable aleatoria
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# Calcular la función de densidad de probabilidad
pdf = norm.pdf(x, mu, sigma)

# Graficar la distribución normal
plt.plot(x, pdf, label='Distribución normal')

# Sombras para el intervalo de interés
x_fill = np.linspace(170, 180, 100)
y_fill = norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill, color='grey', alpha=0.5)

# Configurar la gráfica
plt.title('Distribución normal')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad de probabilidad')
plt.legend()

# Mostrar la gráfica
plt.show()
```

## Caso 3
Una de las distribuciones continuas más conocidas es la distribución uniforme, que describe una variable aleatoria que tiene igual probabilidad de tomar cualquier valor dentro de un rango determinado.

Por ejemplo, supongamos que lanzamos un dado justo, es decir, que tiene igual probabilidad de mostrar cualquier número del 1 al 6. Si definimos la variable aleatoria X como el número obtenido en el lanzamiento del dado, entonces X sigue una distribución uniforme en el intervalo [1, 6].

Donde a y b son los límites inferior y superior del intervalo.

Podemos graficar esta distribución uniforme en Python utilizando la biblioteca Matplotlib de la siguiente manera:

```python
# Definir los límites del intervalo
a = 1
b = 6

# Crear un vector con valores dentro del intervalo
x = np.linspace(a, b, num=100)

# Calcular la función de densidad de probabilidad
f = np.ones_like(x) / (b - a)

# Graficar la función de densidad de probabilidad
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim([0,1])
plt.title('Distribución uniforme en el intervalo [1, 6]')
plt.fill_between(x, f, color='grey', alpha=0.5)
plt.show()
```

Por ejemplo, si queremos calcular la probabilidad de que el tiempo de espera entre dos clientes consecutivos sea menor o igual a 5 minutos, y sabemos que la tasa promedio de llegada de clientes es de 0.2 clientes por minuto, entonces podemos calcularlo en Python de la siguiente manera:
