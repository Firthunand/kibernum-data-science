# Guía Autoestudio: Bondad del Modelo Regresivo

La bondad del modelo en el contexto de la regresión lineal se refiere a la capacidad del modelo para ajustarse adecuadamente a los datos y predecir respuestas futuras de manera precisa. En otras palabras, la bondad del modelo evalúa qué tan bien el modelo explica la variabilidad observada en los datos y cuán confiables son las predicciones realizadas por el modelo.

Una de las medidas más comunes para evaluar la bondad del modelo en regresión lineal es el coeficiente de determinación ($R^{2}$). El coeficiente de determinación proporciona una medida de la proporción de la variabilidad en la variable de respuesta que es explicada por el modelo. Además del $R^{2}$, otras medidas como el error estándar de la regresión (SER), el error cuadrático medio (MSE) y el error absoluto medio (MAE) también se pueden utilizar para evaluar la bondad del modelo. Estas medidas proporcionan información sobre la precisión de las predicciones realizadas por el modelo y pueden ser útiles para comparar diferentes modelos entre sí.

En resumen, la bondad del modelo en regresión lineal se evalúa mediante diversas medidas estadísticas que proporcionan información sobre la capacidad del modelo para explicar la variabilidad en los datos y realizar predicciones precisas. Un buen modelo de regresión lineal debería tener un $R^{2}$ alto y errores de predicción bajos.

**Mean Absolute Error (MAE)**
El Mean Absolute Error (MAE), o Error Absoluto Medio en español, es una medida de la precisión de un modelo de regresión que calcula el promedio de las diferencias absolutas entre los valores observados y los valores predichos por el modelo. Se calcula utilizando la siguiente fórmula:

$MAE = \frac{1}{n}\sum_{i=1}^{n}\left | y_{i}-\widehat{y}_{i} \right |$

Donde,
- $n$ es el número total de observaciones.
- $y_{i}$  son los valores observados.
- $\widehat{y}_{i}$ son los valores predichos por el modelo para la observación 
- $\left | x \right |$ representa el valor absoluto de $x$.

El MAE proporciona una medida de la magnitud promedio de los errores de predicción en el modelo. Cuanto menor sea el MAE, mejor será el rendimiento del modelo, ya que indica que las predicciones están más cerca de los valores observados en promedio. El MAE es útil porque es fácil de interpretar y se expresa en la misma unidad que la variable de respuesta, lo que facilita su comprensión.

**Mean Squared Error (MSE)**

El MSE, del inglés "Mean Squared Error", o Error Cuadrático Medio en español, es una medida de la dispersión de los errores de predicción en un modelo de regresión. Se calcula como el promedio de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo. La fórmula para calcular el MSE es:

$MSE = \frac{1}{n}\sum_{i=1}^{n}(y_{i}-\widehat{y}_{i})^{2}$

Donde,

- $n$ es el número total de observaciones.
- $y_{i}$ son los valores observados.
- $\widehat{y}_{i}$ son los valores predichos por el modelo para la observación $i$.

El MSE proporciona una medida de la calidad de las predicciones realizadas por el modelo. Cuanto menor sea el MSE, mejor será el rendimiento del modelo, ya que indica que los valores predichos están más cerca de los valores observados. Por lo tanto, el MSE se utiliza comúnmente como una métrica para comparar diferentes modelos de regresión y evaluar su capacidad para hacer predicciones precisas.

**Root Mean Squared Error (RMSE)**

El Root Mean Squared Error (RMSE), o Error Cuadrático Medio de la Raíz en español, es otra medida de la precisión de un modelo de regresión. Se calcula como la raíz cuadrada del promedio de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo. La fórmula para calcular el RMSE es:

$RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}\left | y_{i}-\widehat{y}_{i} \right |}$

Donde,
- $n$ es el número total de observaciones.
- $y_{i}$ son los valores observados.
- $\widehat{y}_{i}$ son los valores predichos por el modelo para la observación $i$.

El RMSE proporciona una medida de la dispersión de los errores de predicción en el modelo. Al igual que el MSE, cuanto menor sea el valor del RMSE, mejor será el rendimiento del modelo, ya que indica que las predicciones están más cerca de los valores observados en promedio.

El RMSE es útil porque se expresa en la misma unidad que la variable de respuesta, lo que facilita su interpretación y comparación con otros modelos. Además, al ser una medida de la raíz cuadrada del error cuadrático medio, penaliza de manera más significativa los grandes errores de predicción en comparación con el MAE.

**Coeficiente de Determinación ($R^{2}$)**

El coeficiente de determinación, comúnmente conocido como $R^{2}$ o R-cuadrado, es una medida estadística que indica cuánta variabilidad en la variable dependiente es explicada por el modelo de regresión. En otras palabras, $R^{2}$ proporciona una medida de la bondad del ajuste del modelo a los datos observados.

$R^{2}$ se calcula como la proporción de la variabilidad explicada por el modelo en relación con la variabilidad total de los datos. Matemáticamente, se define como:


$R^{2}  = 1 - \frac{SS_{res}}{SS_{tot}}$

En donde,

- $SS_{res}$ es la suma de los cuadrados de los residuos, es decir, la suma de las diferencias al cuadrado entre los valores observados y los valores predichos por el modelo.
- $SS_{tot}$ es la suma total de los cuadrados, es decir, la suma de las diferencias al cuadrado entre los valores observados y la media de los valores observados.

Es decir,

$SS_{res} = \sum_{i=1}^{n}(y_{i}-\widehat{y}_{i})^{2}$

y

$SS_{tot} = \sum_{i=1}^{n}(y_{i}-\overline{y}_{i})^{2}$


El valor de $R^{2}$ puede variar entre 0 y 1. Un valor de $R^{2}$ cercano a 1 indica que el modelo explica una gran proporción de la variabilidad en los datos, lo que sugiere un buen ajuste del modelo. Por otro lado, un valor de $R^{2}$ cercano a 0 indica que el modelo no explica bien la variabilidad observada en los datos.

En resumen, el coeficiente de determinación ($R^{2}$) es una medida importante de la bondad del ajuste del modelo en la regresión, ya que proporciona información sobre cuánta variabilidad en la variable dependiente es explicada por el modelo en relación con la variabilidad total de los datos.

**Gráfico Q-Q (quantile-quantile)**

Este gráfico compara los cuantiles de los residuos con los cuantiles teóricos de una distribución normal. Si los residuos siguen una distribución normal, los puntos en el gráfico Q-Q deberían formar una línea recta. Desviaciones de esta línea pueden indicar que los residuos no se distribuyen normalmente.

Se recomienda ver el siguiente video para un mayor detalle:
https://www.youtube.com/watch?v=MuJvecsHkc8


```python
datos_observados = y_true - y_pred

# Calcula los cuantiles teóricos de una distribución normal
cuantiles_teoricos = np.random.normal(loc=0, scale=1, size=datos_observados.shape[0])

# Ordena los datos observados y los cuantiles teóricos
datos_observados_sorted = np.sort(datos_observados)

# estandarizamos los datos observados para poder compararlo con los cuantiles teoricos
datos_observados_estandar = (datos_observados_sorted - np.mean(datos_observados_sorted)) / np.std(datos_observados_sorted)


cuantiles_teoricos_sorted = np.sort(cuantiles_teoricos)

# Crea el gráfico Q-Q
plt.figure(figsize=(8, 6))
plt.scatter(cuantiles_teoricos_sorted, datos_observados_estandar, color='blue', alpha=0.6)
plt.xlabel('Cuantiles teóricos')
plt.ylabel('Cuantiles observados')
plt.title('Gráfico Q-Q')
plt.plot([-3, 3], [-3, 3], color='red', linestyle='--')  # Línea de referencia para distribución normal
plt.grid(True)
plt.show()

```
